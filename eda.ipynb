```python
from dataset_loader import load_dataset_amazon_test
import matplotlib.pyplot as plt

# Reset pyplot
plt.clf()
plt.close('all')
plt.style.use("dark_background")

# Load Test Dataset
df = load_dataset_amazon_test()
print("Size of Amazon Test Dataset:", df.size)
```

```python
df["text_full"] = (df["title"].astype(str) + " " + df["comment"].astype(str)).str.strip()
df["length"] = df["text_full"].str.len()

# Plot histogram
plt.figure(figsize=(10,5))
plt.hist(df["length"], bins=50)
plt.title("Review Length Distribution")
plt.xlabel("Number of Characters")
plt.ylabel("Count")
plt.show()
```

```python
# compute text length
df["length"] = df["text_full"].str.len()

# separate by class
pos = df[df["label"] == 2]["length"]
neg = df[df["label"] == 1]["length"]

# plot
plt.figure(figsize=(10,5))
plt.hist(pos, bins=50, alpha=0.5, label="Positive (label=2)")
plt.hist(neg, bins=50, alpha=0.5, label="Negative (label=1)")
plt.xlabel("Number of Characters")
plt.ylabel("Count")
plt.title("Review Length Distribution by Sentiment")
plt.legend()
plt.show()
```
